### github repo contributors

Fetch the list of contributors in the repository 'CodingNinjasCodes/JSNotes' in Github from Github API using OAuth(access token).
Print the username of contributors and its number of commit done.

```python
import requests
import json
p=requests.get('https://api.github.com/repos/CodingNinjasCodes/JSNotes/stats/contributors',headers={'token':'ae57026eceb0a87d5349e2534c8f1d102b8378b6'})
data=p.json()
dic=dict()
for i in data:
    name =i['author']['login']
    if name in dic:
        dic[name]+=i['total']
    else:
        dic[name]=i['total']
for row in dic:
    print(row,dic[row])
```
